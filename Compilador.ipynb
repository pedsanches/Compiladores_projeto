```python
def insere_Token_recebido(token_rec):
  if(token_rec not in tabela_simbolos):
    tabela_simbolos.append(token_rec)
  return

print(tabela_simbolos)

insere_Token_recebido(Token("teste","teste","teste"))

print(tabela_simbolos)
```

```python
def procura_Lexema_recebido(lexema_rec):
  for i in range(len(tabela_simbolos)):
    if(tabela_simbolos[i].Lexema == lexema_rec):
      return True
  return False

def procura_Token_recebido(token_rec):
  return token_rec in tabela_simbolos

print (procura_Token_recebido("teste"))
```

```python
def atualiza_Token_recebido(token_rec, token_atualizado):
  if(token_rec in tabela_simbolos):
    tabela_simbolos[tabela_simbolos.index(token_rec)] = token_atualizado
  else:
    print("Token não presente na tabela!")

  
atualiza_Token_recebido(Token("teste","teste","teste"),Token("teste","teste","teste123"))
print(tabela_simbolos)
```

```python
def retorna_token_existe_tabela_lexema(lexema_rec):
   if(procura_Lexema_recebido(lexema_rec)):
     for i in tabela_simbolos:
       if(i.Lexema == lexema_rec):
         return i
        

print(retorna_token_existe_tabela_lexema("teste"))
```

##Scanner

```python
conteudos = []

def atualiza_arquivo():
  global conteudos
  with open('/content/sample_data/Teste.arg') as f:
    conteudos = f.readlines()

cursor = 0
linha = 0

def zera():
  global cursor
  global linha 

  cursor = linha = 0
```

```python
def Ignorar():
  return

def Vir():
  return Token("vir",",","NULO")

def ERRO(op):
  if(op == 1):
    print("ERRO LÉXICO – Caractere inválido na linguagem, linha: ",linha, ", coluna:",cursor)
  elif(op == 2):
    print("ERRO LÉXICO – Caractere inesperado para a situação, linha: ",linha, ", coluna:",cursor)
  elif(op == 3):
    print("ERRO LÉXICO – Falta de fechamento do literal: ",linha, ", coluna:",cursor)
  return Token("erro","ERRO","NULO")

def PT_V():
  return Token("pt_v",";","NULO")

def FC_P():
  return Token("fc_p",")","NULO")

def AB_P():
  return Token("ab_p","(","NULO")

def OPM(op):
  if(op == "+"):
    return Token("opm","+","NULO")
  elif(op == "-"):
    return Token("opm","-","NULO")
  elif(op == "*"):
    return Token("opm","*","NULO")
  else:
    return Token("opm","/","NULO")

def RCB():
  return Token("rcb","<-","NULO")

def OPR(op):
  if(op == "<"):
    return Token("opr","<","NULO")
  elif(op == ">"):
    return Token("opr",">","NULO")
  elif(op == ">="):
    return Token("opr",">=","NULO")
  elif(op == "<="):
    return Token("opr","<=","NULO")
  elif(op == "="):
    return Token("opr","=","NULO")
  else:
    return Token("opr","<>","NULO")

def EOF():
  return Token("$","EOF","NULO")

def coment():
  return

def id(op):
  if(procura_Lexema_recebido(op)):
    for i in tabela_simbolos:
      if(i.Lexema == op):
        return i
  else:
    aux = Token("id",op,"NULO")
    insere_Token_recebido(aux)
    return aux

def Lit(op):
  return Token("lit",op,"literal")

def NumI(op):
  return Token("num",op,"inteiro")

def NumR(op):
  return Token("num",op,"real")




```

#Reconhecedores

```python
def reconhecedor_id(op):
  global cursor
  global linha

  cursor += 1
  
  if(cursor == len(conteudos[linha])-1 and conteudos[linha][-1] != " " and conteudos[linha][-1] != "\n" and conteudos[linha][-1] != "\t"):
    return id(op+conteudos[linha][-1])

  elif((conteudos[linha][cursor] in letras) or (conteudos[linha][cursor] == "_")):
    return reconhecedor_id(op+conteudos[linha][cursor])

  elif(conteudos[linha][cursor] in demais):
    cursor -= 1
    return id(op)

  else:
    if(conteudos[linha][cursor] not in possiveis):
      return ERRO(1)
    return ERRO(2)
```

```python
def reconhecedor_literal(op):
  global cursor
  global linha

  cursor += 1

  if(conteudos[linha][cursor] not in possiveis):
    return ERRO(1)

  if(conteudos[linha][cursor] in possiveis and conteudos[linha][cursor] != "\""):
    return reconhecedor_literal(op+conteudos[linha][cursor])
  elif(conteudos[linha][cursor] == "\""):
    return Lit(op+conteudos[linha][cursor])
  elif(len(conteudos[linha])-1 == cursor and conteudos[linha][cursor] != "\""):
    return ERRO(3)
  else:
    return ERRO(2)
```

```python
def reconhecedor_op_relacional_atribuicao(op):
  global cursor
  global linha

  if(op == ">" or op == "<"):
    if(cursor < len(conteudos[linha])-1):
      if(conteudos[linha][cursor+1] == "="):
        cursor += 1
        return OPR(op + conteudos[linha][cursor])
      if(op == "<"):
        if(conteudos[linha][cursor+1] == ">"):
          cursor += 1
          return OPR(op + conteudos[linha][cursor])
        elif(conteudos[linha][cursor+1] == "-"):
          cursor += 1
          return RCB()
  if(conteudos[linha][cursor] not in possiveis):
    return ERRO(1)
  return OPR(op)
```

```python
def reconhecedor_comentario():
  global cursor
  global linha

  cursor += 1
  while(conteudos[linha][cursor] != "}"):
    
    if(conteudos[linha][cursor] == conteudos[linha][-1]):
      linha += 1
      cursor = 0
    else:
      cursor += 1
    
  if(conteudos[linha][cursor] not in possiveis):
    return ERRO(1)
```

```python
def reconhecedor_numerico(op):
  global cursor
  global linha

  real = False

  if(len(conteudos[linha])-1>cursor):
     cursor += 1
  else:
    if(real):
      return NumR(op)
    else:
      return NumI(op) 

  while(conteudos[linha][cursor] in digitos):
    op = op + conteudos[linha][cursor] 
    if(cursor == len(conteudos[linha])-1):
      break
    cursor += 1
    
  if(cursor == len(conteudos[linha])-1 and linha == len(conteudos)-1):
    if(real):
      return NumR(op)
    else:
      return NumI(op)

  if(conteudos[linha][cursor]  == "."):
    real = True
    op = op + conteudos[linha][cursor] 
    cursor += 1
    if(conteudos[linha][cursor] not in digitos):
      return ERRO(2)
    else:
      while(conteudos[linha][cursor] in digitos):
        op = op + conteudos[linha][cursor] 
        if(cursor == len(conteudos[linha])-1):
          break
        cursor += 1
  
    if(cursor == len(conteudos[linha])-1 and linha == len(conteudos)-1):
      if(real):
        return NumR(op)
      else:
       return NumI(op)

  if(conteudos[linha][cursor]  == "e" or conteudos[linha][cursor]  == "E"):
    op = op + conteudos[linha][cursor] 
    cursor += 1

    if(conteudos[linha][cursor]  == "+"):
      op = op + conteudos[linha][cursor] 
      cursor += 1
    elif(conteudos[linha][cursor]  == "-"):
      real = True
      op = op + conteudos[linha][cursor] 
      cursor += 1

    if(conteudos[linha][cursor] not in digitos):
      return ERRO(2)
    else:
      while(conteudos[linha][cursor] in digitos):
        op = op + conteudos[linha][cursor] 
        if(cursor == len(conteudos[linha])-1):
          break
        cursor += 1
  
  if(conteudos[linha][cursor] not in possiveis):
    return ERRO(1)

  if(real):
    return NumR(op)
  else:
    return NumI(op)
    
```

#Cursor do scanner

```python

def scanner():
  global cursor
  global linha

  while(1):
    if(linha >= len(conteudos)):
      #print(linha, len(conteudos))
      return EOF()

    atual = conteudos[linha][cursor]
    #print(atual)
    token_retorno = None
    estado = False

    if(conteudos[linha][cursor] == " " or conteudos[linha][cursor] == "\n" or conteudos[linha][cursor] == "\t"):
      pass

    #reconhecer os ids
    elif(conteudos[linha][cursor] in letras):
      token_retorno = reconhecedor_id(conteudos[linha][cursor])
      estado = True

    #reconhecer os literais
    elif(conteudos[linha][cursor] == "\""):
      token_retorno = reconhecedor_literal(conteudos[linha][cursor]) 
      estado = True

    #reconhecer os Operadores relacionais ou Atribuição
    elif(conteudos[linha][cursor] == "=" or conteudos[linha][cursor] == ">" or conteudos[linha][cursor] == "<"):
      token_retorno = reconhecedor_op_relacional_atribuicao(conteudos[linha][cursor])
      estado = True

    #reconhecer os Operadores Aritmeticos
    elif(conteudos[linha][cursor] == "+" or conteudos[linha][cursor] == "-" or conteudos[linha][cursor] == "*" or conteudos[linha][cursor] == "/" ):
      token_retorno = OPM(conteudos[linha][cursor])
      estado = True

    #reconhecer abre parenteses   
    elif(conteudos[linha][cursor] == "(" ):
      token_retorno = AB_P()
      estado = True

    #reconhecer fecha parenteses   
    elif(conteudos[linha][cursor] == ")" ):
      token_retorno = FC_P()
      estado = True

    #reconhecer ponto e virgula
    elif(conteudos[linha][cursor] == ";" ):
      token_retorno = PT_V()
      estado = True  

    #reconhecer virgula
    elif(conteudos[linha][cursor] == "," ):
      token_retorno = Vir()
      estado = True

    #reconhecedor de comentário
    elif(conteudos[linha][cursor] == "{" ):
      reconhecedor_comentario()

    elif(conteudos[linha][cursor] in digitos ):
      token_retorno = reconhecedor_numerico(conteudos[linha][cursor])
      cursor-=1
      estado = True

    elif(linha == len(conteudos) and cursor == len(conteudos[linha])):
      token_retorno = EOF()
      estado = True

    else:
      token_retorno = ERRO(1)
      estado = True

    if(cursor == len(conteudos[linha])-1):
      linha += 1
      cursor = 0
      if(estado):
        return token_retorno
    else:
      cursor += 1
      if(estado):
        return token_retorno

     

```
